```python
import pandas as pd
import numpy as np

customers = pd.read_csv('data/olist_customers_dataset.csv')
orders = pd.read_csv('data/olist_orders_dataset.csv')
order_items = pd.read_csv('data/olist_order_items_dataset.csv')
order_payments = pd.read_csv('data/olist_order_payments_dataset.csv')
order_reviews = pd.read_csv('data/olist_order_reviews_dataset.csv')
products = pd.read_csv('data/olist_products_dataset.csv')
sellers = pd.read_csv('data/olist_sellers_dataset.csv')
category_translation = pd.read_csv('data/product_category_name_translation.csv')

print("✅ Data loaded")
```

```python
print("Missing values di products sebelum cleaning:")
print(products.isnull().sum())
```

```python
print("Missing values di products sebelum cleaning:")
print(products.isnull().sum())

products['product_category_name'] = products['product_category_name'].fillna('unknown')

products = products.merge(
    category_translation, 
    on='product_category_name', 
    how='left'
)

products['product_category_name_english'] = products['product_category_name_english'].fillna('unknown')

print("✅ Products cleaned")
print("\nMissing values setelah cleaning:")
print(products.isnull().sum())
```

```python
print("Missing values di orders sebelum cleaning:")
print(orders.isnull().sum())
```

```python
print("\n" + "="*50)
print("Cleaning Orders...")
print("="*50)

print("Missing values sebelum cleaning:")
print(orders.isnull().sum())

orders['order_approved_at'] = orders['order_approved_at'].fillna('not_approved')
orders['order_delivered_carrier_date'] = orders['order_delivered_carrier_date'].fillna('not_delivered_to_carrier')
orders['order_delivered_customer_date'] = orders['order_delivered_customer_date'].fillna('not_delivered')

date_columns = [
    'order_purchase_timestamp',
    'order_approved_at', 
    'order_delivered_carrier_date',
    'order_delivered_customer_date',
    'order_estimated_delivery_date'
]

for col in date_columns:
    if col in orders.columns:
        orders[col] = pd.to_datetime(orders[col], errors='coerce')

print("✅ Orders cleaned")
```

```python
print("Missing values di order_reviews sebelum cleaning:")
print(order_reviews.isnull().sum())
```

```python
print("\n" + "="*50)
print("Cleaning Order Reviews...")
print("="*50)

print("Missing values sebelum cleaning:")
print(order_reviews.isnull().sum())

order_reviews['review_comment_title'] = order_reviews['review_comment_title'].fillna('no_comment')
order_reviews['review_comment_message'] = order_reviews['review_comment_message'].fillna('no_comment')

print("✅ Order reviews cleaned")
```

```python
date_columns = [
    'order_purchase_timestamp',
    'order_approved_at', 
    'order_delivered_carrier_date',
    'order_delivered_customer_date',
    'order_estimated_delivery_date'
]

for col in date_columns:
    if col in orders.columns:
        orders[col] = pd.to_datetime(orders[col], errors='coerce')

print("✅ Date columns converted")
```

```python
print("DUPLICATE CHECK:")
print("="*50)
print(f"customers: {customers.duplicated().sum()} duplicate rows")
print(f"orders: {orders.duplicated().sum()} duplicate rows")
print(f"order_items: {order_items.duplicated().sum()} duplicate rows")
print(f"order_payments: {order_payments.duplicated().sum()} duplicate rows")
print(f"order_reviews: {order_reviews.duplicated().sum()} duplicate rows")
print(f"products: {products.duplicated().sum()} duplicate rows")
```

```python
customers.to_csv('data/cleaned_customers.csv', index=False)
orders.to_csv('data/cleaned_orders.csv', index=False)
order_items.to_csv('data/cleaned_order_items.csv', index=False)
order_payments.to_csv('data/cleaned_order_payments.csv', index=False)
order_reviews.to_csv('data/cleaned_order_reviews.csv', index=False)
products.to_csv('data/cleaned_products.csv', index=False)

print("✅ Semua cleaned data telah disimpan!")
```
