```python
import pandas as pd
import sqlite3
import os

print("Loading cleaned data...")
print("=" * 40)

customers = pd.read_csv('data/cleaned_customers.csv')
orders = pd.read_csv('data/cleaned_orders.csv')
order_items = pd.read_csv('data/cleaned_order_items.csv')
order_payments = pd.read_csv('data/cleaned_order_payments.csv')
order_reviews = pd.read_csv('data/cleaned_order_reviews.csv')
products = pd.read_csv('data/cleaned_products.csv')
sellers = pd.read_csv('data/olist_sellers_dataset.csv')

print("✅ Data loaded:")
print(f"   customers: {customers.shape}")
print(f"   orders: {orders.shape}")
print(f"   order_items: {order_items.shape}")
print(f"   order_payments: {order_payments.shape}")
print(f"   order_reviews: {order_reviews.shape}")
print(f"   products: {products.shape}")
print(f"   sellers: {sellers.shape}")
```

```python
conn = sqlite3.connect('data/olist.db')
print("✅ Database created: data/olist.db")
```

```python
print("Loading data to SQLite...")
print("=" * 40)

customers.to_sql('customers', conn, if_exists='replace', index=False)
print("✅ customers table created")

orders.to_sql('orders', conn, if_exists='replace', index=False)
print("✅ orders table created")

order_items.to_sql('order_items', conn, if_exists='replace', index=False)
print("✅ order_items table created")

order_payments.to_sql('order_payments', conn, if_exists='replace', index=False)
print("✅ order_payments table created")

order_reviews.to_sql('order_reviews', conn, if_exists='replace', index=False)
print("✅ order_reviews table created")

products.to_sql('products', conn, if_exists='replace', index=False)
print("✅ products table created")

sellers.to_sql('sellers', conn, if_exists='replace', index=False)
print("✅ sellers table created")

print("=" * 40)
print("✅ Semua data berhasil masuk ke database!")
```

```python
tables = pd.read_sql("SELECT name FROM sqlite_master WHERE type='table';", conn)
print("TABLES IN DATABASE:")
print(tables)
```

```python
print("ROW COUNTS:")
print("=" * 40)

query_counts = """
SELECT 'customers' as table_name, COUNT(*) as count FROM customers
UNION ALL
SELECT 'orders', COUNT(*) FROM orders
UNION ALL
SELECT 'order_items', COUNT(*) FROM order_items
UNION ALL
SELECT 'order_payments', COUNT(*) FROM order_payments
UNION ALL
SELECT 'order_reviews', COUNT(*) FROM order_reviews
UNION ALL
SELECT 'products', COUNT(*) FROM products
UNION ALL
SELECT 'sellers', COUNT(*) FROM sellers;
"""

counts = pd.read_sql(query_counts, conn)
print(counts)
```

```python
conn.close()
print("✅ Database connection closed")
```
